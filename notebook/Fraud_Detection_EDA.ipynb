````python
# Import data visualization libraries
import matplotlib.pyplot as plt  # Primary plotting interface (MATLAB-style)
import seaborn as sns  # High-level statistical visualization (based on matplotlib)
import matplotlib.patches as mpatches  # For creating custom legend handles (colored boxes, circles, etc.)

# Magic command for Jupyter notebooks - displays plots inline
%matplotlib inline
````

````python
# TODO: Complete the method below for plotting data distributions
def plot_numeric_distributions(df: pd.DataFrame, numeric_cols: list):
    """Visualize distribution of numerical features split by fraud class.
    
    Args:
        df: DataFrame containing the data
        numeric_cols: List of numerical column names to plot
    """
    df = df.copy()  # Create a copy to avoid modifying original DataFrame

    # Color and appearance settings
    fraud_labels = ['Legitimate (0)', 'Fraud (1)']  # Legend labels
    colors = sns.color_palette('Set1', n_colors=2)  # 2 distinct colors for the classes
    
    # Chart global settings
    plt.figure(figsize=(20, 15))  # Set large figure size
    plt.suptitle('Distribution of numerical features based on fraud class', fontsize=18, y=1.02)

    # Calculate grid dimensions for subplots
    n_cols = 3  # Number of columns in subplot grid
    n_rows = (len(numeric_cols) // n_cols) + 1  # Calculate required rows

    # Plot grouped histograms for each feature
    for i, col in enumerate(numeric_cols, 1):
        plt.subplot(n_rows, n_cols, i)  # Create subplot
        sns.histplot(
            data=df,
            x=col,
            hue='Fraud_Label',  # Split by fraud class
            bins=20,  # Number of histogram bins
            palette=colors,  # Use our color scheme
            alpha=0.6,  # Transparency level
            kde=True  # Add kernel density estimate
        )
        plt.title(f'Distribution {col}', fontsize=12)
        plt.xlabel(col, fontsize=10)
        plt.ylabel('Number', fontsize=10)

    # Add custom legend
    patches = [mpatches.Patch(color=colors[i], label=fraud_labels[i]) for i in range(2)]
    plt.legend(
        handles=patches,
        title='Transaction status',
        bbox_to_anchor=(1.05, 1),  # Position outside plot
        loc='upper left'
    )

    plt.tight_layout()  # Adjust spacing between subplots
    plt.show()
````

````python
# Get numerical columns (excluding label)
numeric_cols = df.select_dtypes(include='number').columns.tolist()
numeric_cols.remove('Fraud_Label')  # Remove the target

# Plot
plot_numeric_distributions(df, numeric_cols)
````

````python
def drop_columns(df: pd.DataFrame, columns: list[str]):
    """Drop specified columns from a pandas DataFrame.
    
    Args:
        df: Input DataFrame to modify
        columns: List of column names to remove
        
    Returns:
        A new DataFrame with specified columns removed
    """
    return df.drop(columns=columns, axis=1)  # axis=1 indicates column-wise operation
````

````python
import numpy as np
from scipy import stats
````

````python
# Select numeric columns (int64 and float64) excluding the target variable 'Fraud_Label'
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_cols = [col for col in numeric_cols if col != "Fraud_Label"]

def detect_outliers(df, numeric_cols):
    """Detect and count outliers in numeric columns using IQR and Z-Score methods.
    
    Args:
        df: Input DataFrame
        numeric_cols: List of numeric column names to analyze
    """
    for col in numeric_cols:
        # TODO: Identify outliers using IQR method
        Q1 = df[col].quantile(0.25)  # First quartile (25th percentile)
        Q3 = df[col].quantile(0.75)  # Third quartile (75th percentile)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR  # Lower threshold for outliers
        upper_bound = Q3 + 1.5 * IQR  # Upper threshold for outliers
        outliers_iqr = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]  # Count of IQR outliers
        
        # TODO: Identify outliers using Z-Score method
        z_scores = np.abs(stats.zscore(df[col]))  # Absolute Z-scores
        outliers_zscore = (z_scores > 3).sum()  # Count of Z-score outliers (beyond 3 std dev)
        
        # Print results for current column
        print(f"Outliers using IQR for column {col}: {outliers_iqr}")
        print(f"Outliers using Z-Score for column {col}: {outliers_zscore}")
        print()  # Blank line for readability

# Call the function to detect and display outliers
detect_outliers(df, numeric_cols)
````

````python
# TODO: Remove outlier data points

def remove_outliers_iqr(df, numeric_cols):
    """Remove outliers from numeric columns using the IQR method.
    
    Args:
        df: Input DataFrame
        numeric_cols: List of numeric column names to process
        
    Returns:
        A new DataFrame with outliers removed
    """
    df_clean = df.copy()  # Create a copy to avoid modifying original DataFrame
    
    for col in numeric_cols:
        # Calculate IQR bounds
        Q1 = df_clean[col].quantile(0.25)  # 25th percentile (first quartile)
        Q3 = df_clean[col].quantile(0.75)  # 75th percentile (third quartile)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR  # Lower cutoff for outliers
        upper_bound = Q3 + 1.5 * IQR  # Upper cutoff for outliers
        
        # Filter rows where values in this column are within bounds (non-outliers)
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    
    return df_clean  # Return cleaned DataFrame

# Apply outlier removal to our DataFrame
df = remove_outliers_iqr(df, numeric_cols)
````

````python
# TODO: Complete the method below
def impute_missing_data(
    df: pd.DataFrame,
    strategy_dict: dict = None,
    auto_impute: bool = True
) -> pd.DataFrame:
    """
    Handle missing values using either automatic or specified strategies
    
    Parameters:
        df (pd.DataFrame): Input DataFrame
        strategy_dict (dict): Dictionary of imputation strategies (optional)
        auto_impute (bool): Use automatic imputation (default: True)
    
    Returns:
        pd.DataFrame: DataFrame without missing values
    """
    df_clean = df.copy()  # Work on copy to avoid modifying original

    # Automatic imputation
    if auto_impute:
        # Fill numeric columns with mean
        num_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
        for col in num_cols:
            df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
        
        # Fill categorical columns with mode
        cat_cols = df_clean.select_dtypes(include=['object', 'category']).columns
        for col in cat_cols:
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
    
    # Custom strategy imputation
    elif not auto_impute and strategy_dict is not None:
        for col, strategy in strategy_dict.items():
            if col not in df_clean.columns:
                print(f"Column '{col}' not found in the DataFrame.")
                continue  # Skip non-existent columns
            
            # Numeric strategies
            if strategy == 'mean':
                df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
                continue
            elif strategy == 'median':
                df_clean[col] = df_clean[col].fillna(df_clean[col].median())
                continue
            # Categorical strategies
            elif strategy == 'mode':
                df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
                continue
            elif strategy == 'unknown':
                df_clean[col] = df_clean[col].fillna('unknown')
                continue
            elif strategy == "drop_na":
                df_clean = df_clean.dropna(subset=[col])
                continue
            # KNN strategy (numeric only)
            elif strategy == 'knn':
                if pd.api.types.is_numeric_dtype(df[col]) and df[col].nunique() >= 10:
                    scaler = StandardScaler()
                    scaled_data = scaler.transform(df_clean[col])
                    imputer = KNNImputer(n_neighbors=3)
                    scaled_data = imputer.transform(scaled_data)
                    df_clean.loc[:, col] = scaler.inverse_transform(scaled_data)

    return df_clean
````

````python
# Custom imputation strategies for each column
custom_strategies = {
    'Transaction_Type': 'mode',  # Fill with most frequent category
    'Device_Type': 'mode',
    'Location': 'unknown',  # Mark missing as 'unknown'
    'Merchant_Category': 'mode',
    'Card_Type': 'mode',
    "Authentication_Method": "mode",
    "Transaction_Month": "mode",  # Temporal features
    "Transaction_Day": "mode",
    "Transaction_Hour": "mode",
    "Transaction_DayOfWeek": "mode",
    "Fraud_Label": "drop_na"  # Remove rows with missing target
}

# TODO: Apply imputation to dataset
df = impute_missing_data(df, strategy_dict=custom_strategies, auto_impute=False)
````

````python
# Import required libraries for imbalanced learning and models
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.combine import SMOTETomek, SMOTEENN
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier
from collections import Counter
from sklearn.metrics import roc_auc_score, classification_report
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from lightgbm import LGBMClassifier
````

````python
def data_balancing(methods: dict, models:dict, X_train_encoded, X_val_encoded, X_test_encoded,
                   y_train_encoded, y_val_encoded, y_test_encoded):
    """Evaluate different sampling methods with various models.
    
    Args:
        methods: Dictionary of sampling methods (SMOTE, ADASYN, etc.)
        models: Dictionary of classification models
        X_train_encoded: Encoded training features
        X_val_encoded: Encoded validation features  
        X_test_encoded: Encoded test features
        y_train_encoded: Training labels
        y_val_encoded: Validation labels
        y_test_encoded: Test labels
        
    Returns:
        Dictionary containing all evaluation results
    """
    
    # Dictionary to store evaluation results
    results = {}
    
    # Print initial class distribution
    print("Initial Y_Train distribution:", (y_train_encoded.shape))
    
    # Evaluate each sampling method
    for method_name, method in methods.items():
        print(f"\n--- Processing {method_name} method ---")
        
        # Apply resampling ONLY to training data
        X_train_resampled, y_train_resampled = method.fit_resample(X_train_encoded, y_train_encoded)
        print(f"New Train distribution ({method_name}):", Counter(y_train_resampled))
        
        # Store results for current method
        results[method_name] = {}
        
        # Evaluate each model with current sampling method
        for model_name, model in models.items():
            print(f"\nTraining {model_name} model...")
            
            # Train model
            model.fit(X_train_resampled, y_train_resampled)
            
            # Get predictions and probabilities
            y_train_pred = model.predict(X_train_resampled)
            y_train_proba = model.predict_proba(X_train_resampled)[:, 1]  # Positive class probability
            y_val_pred = model.predict(X_val_encoded)
            y_val_proba = model.predict_proba(X_val_encoded)[:, 1]
            y_test_pred = model.predict(X_test_encoded)
            y_test_proba = model.predict_proba(X_test_encoded)[:, 1]
            
            # Print evaluation metrics
            print(f"""
                 ------------- Train Evaluation Metrics: {model_name} + {method_name} ---------------------- 
                'AUC-ROC': {roc_auc_score(y_train_resampled, y_train_proba)},
                <---classification_report--->
                {classification_report(y_train_resampled, y_train_pred)}
                

                 ----------- Validation Evaluation Metrics: {model_name} + {method_name} -------------
                'AUC-ROC': {roc_auc_score(y_val_encoded, y_val_proba)},
                <---classification_report--->
                {classification_report(y_val_encoded, y_val_pred)}
                
                -------------- Test Evaluation Metrics: {model_name} + {method_name} --------------------- 
                'AUC-ROC': {roc_auc_score(y_test_encoded, y_test_proba)},

                <---classification_report--->
                {classification_report(y_test_encoded, y_test_pred)}
__________________________________________________________________________________________________________________________
            """)

    return results
````
